### Q1. Which function is used to open a file? What are the different modes of opening a file? Explain each mode of file opening.

### Answer = We use "open" function to open a file. 

### Answer = There are different modes of opening a file are: (1) 'r' (2) 'rb' (3) 'r+' (4) 'w' (5) 'wb' (6) 'w+' (7) 'a' (8) 'ab' (9) 'a+' (10) 'x'

### (1) 'r'  = It's opens the file for reading only.
### (2) 'rb' = It opens the file for reading only in binary format.
### (3) 'r+' = It opens the file reading and writing both.
### (4) 'w'  = It opens the file to write only
### (5) 'wb' = It opens the file to write only in binary format.
### (6) 'w+' = It opens the file to write and read both 
### (7) 'a'  = It will append the existing file.
### (8) 'ab' = It opens the file in the append mode in binary format.
### (9) 'a+' = It opens a file to append and read both .
### (10) 'x' = Exclusive creation mode. Opens the file for writing, but only if it doesn't already exist. If the file exists, a FileExistsError is raised.

### Q4. Explain the following with python code: read(), readline() and readlines().

```python

'''
read():
The read() method is used to read the entire contents of a file as a single string. It returns the contents of the file as a string. Here's an example:
'''
r = open("File.txt", 'r')
print(r.read())
r.close()
```

```python
'''
readline():
The readline() method is used to read a single line from the file. It returns the current line as a string and moves the file pointer to the next line. Here's an example:
'''
r = open("File.txt", 'r')
print(r.readline())
r.close()
```

```python
'''
readlines():
The readlines() method is used to read all lines from the file and return them as a list of strings. Each line in the file becomes an element in the list. Here's an example:
'''
r = open("File.txt", 'r')
print(r.readlines())
r.close()
```

### Q5. Explain why with statement is used with open(). What is the advantage of using with statement and open() together?

### Answer =  The with statement in Python is used with the open() function to provide a cleaner and more concise way of handling file operations. It ensures that the file is automatically closed, regardless of whether an exception occurs or not, leading to improved resource management and exception handling.

### Advantage of Using With Statement and open() Together:
Advantages of using the with statement with open() include:

Automatic Resource Management: The with statement automatically takes care of opening the file and guarantees that it will be closed properly when the block of code is exited. This eliminates the need to manually call the close() method, reducing the chance of forgetting to close the file and potential resource leaks. It promotes good programming practices by ensuring that resources are managed efficiently.

Exception Safety: The with statement ensures that the file is closed even if an exception occurs within the block of code. It provides a built-in mechanism to handle exceptions and guarantees that the file will be closed before the exception is propagated further. This helps maintain data integrity and prevents leaving the file in an invalid or inconsistent state.

Code Readability and Simplicity: By using the with statement, the code becomes more readable and concise. The intent of opening and closing the file is clearly indicated by the with block. It eliminates the need for separate try-except-finally blocks or manually managing the closing of the file, resulting in cleaner and more maintainable code.

Overall, the with statement in combination with the open() function simplifies file handling, promotes good resource management, ensures exception safety, and improves the readability of your code. It is considered the recommended approach for working with files in Python.

### Q6. Explain the write() and writelines() functions. Give a suitable example.

### Write() =  The write() function writes a fixed sequence of character to a file. The write() does not actually write data to a file but to a buffer. When the close() function is called , it is written in file.

```python
# Example of Write Function
file_object = open('file.txt', 'w')
file_object.write("Hello, World!")
file_object.close()
```

### Writelines(): The writelines writes a sequence of the strings to a file . It's syntax is : file.writelines(seq).
### So Whenever we have to write a sequence of string, We will use writelines(), instead of write().

```python
file_object = open('file.txt', 'w')
lines = ['Line 1\n', 'Line 2\n', 'Line 3\n']
file_object.writelines(lines)
file_object.close()
```
